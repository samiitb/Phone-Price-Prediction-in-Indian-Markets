BEST PHONE IN EVERY PRICE CATEGORY

```python
# Group the data by price category
grouped_data = data.groupby(pd.cut(data['Price'], bins=[0, 20000, 40000, 60000, float('inf')]))

# Iterate over each price category and select the best phone
best_phones = []
for category, group in grouped_data:
    best_phone = group.loc[group['RAM (MB)'].idxmax()]  # Select the phone with the highest RAM
    best_phones.append(best_phone)

# Create a DataFrame of the best phones in each price category
best_phones_df = pd.DataFrame(best_phones)

# Visualize the best phones in each price category
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Battery capacity vs. price
axes[0, 0].scatter(data['Battery capacity (mAh)'], data['Price'])
axes[0, 0].set_xlabel('Battery capacity (mAh)')
axes[0, 0].set_ylabel('Price')
axes[0, 0].set_title('Battery Capacity vs. Price')

# Screen size vs. price
axes[0, 1].scatter(data['Screen size (inches)'], data['Price'])
axes[0, 1].set_xlabel('Screen size (inches)')
axes[0, 1].set_ylabel('Price')
axes[0, 1].set_title('Screen Size vs. Price')

# Bar plot of operating system counts
os_counts = data['Operating system'].value_counts()
axes[1, 0].bar(os_counts.index, os_counts.values)
axes[1, 0].set_xlabel('Operating System')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_title('Operating System Distribution')
# Price category vs. best phones
price_categories = ['0-20k', '20k-40k', '40k-60k', '60k+']
best_phones_df['Price Category'] = pd.cut(best_phones_df['Price'], bins=[0, 20000, 40000, 60000, float('inf')], labels=price_categories)
price_counts = best_phones_df['Price Category'].value_counts()
axes[1, 1].bar(price_counts.index, price_counts.values)
axes[1, 1].set_xlabel('Price Category')
axes[1, 1].set_ylabel('Count')
axes[1, 1].set_title('Best Phone Count in Each Price Category')

plt.tight_layout()
plt.show()
# Display the best phones in each price category
print(best_phones_df[['Price Category', 'Brand', 'Model', 'RAM (MB)', 'Battery capacity (mAh)', 'Screen size (inches)']])
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
df = pd.read_csv('/content/ndtv_data_final.csv')
```

```python
# Specify the target variable and predictor(s)
target_variable = 'Price'
predictors = ['Battery capacity (mAh)', 'RAM (MB)', 'Internal storage (GB)',]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[predictors], df[target_variable], test_size=0.2, random_state=42)
```

```python
# Create a linear regression object
regression_model = LinearRegression()

# Fit the model to the training data
regression_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = regression_model.predict(X_test)

# Calculate the mean squared error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate the coefficient of determination (R-squared)
r2 = r2_score(y_test, y_pred)
```

```python
# Print the evaluation metrics
print('Mean Squared Error (MSE):', mse)
print('Coefficient of Determination (R-squared):', r2)
```
